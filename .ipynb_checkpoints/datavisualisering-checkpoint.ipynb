```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

df = pd.read_csv('kategorisert_data.csv', sep = ";")
del df['Tidsmerke']
print(df.columns)
```

```python
#Heatmap for √• se korrelasjon mellom variable seg imellom
cols = ['Kj√∏nn', 'Alder', 'Utdanning', 'frekvens', 'helsenorge_tiltro',
       'facebook_tiltro', 'mainstream_tiltro', 'faktisk_tiltro',
       'resett_tiltro', 'fhi_tiltro']
#'helsenorge1', 'fb1', 'mainstream1', 'faktisk1', 'resett1', 'fhi1','Viktig'
qf = df[cols]
correlation = qf.corr()
hm = correlation
labels = ['Alder', 'Utdanning', 'Oppdatert', 'Helsenorge',
       'Facebook', 'Mainstream', 'Faktisk.no',
       'Resett', 'FHI']

fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(hm, annot=False, cmap='Blues', xticklabels=[], yticklabels=[])
sns.set(font_scale=1.1)
sns.heatmap(hm, mask=hm < 0.1, cbar=False,
            annot=True, annot_kws={"weight": "bold"}, fmt=".2", alpha=0.0, yticklabels=labels, xticklabels=labels)
```

```python
q1=df['enklere_forklart']
labels={0:'Uenig', 1:'Litt uenig', 2:'Vet ikke', 3:'Litt enig', 4:'Enig'}
q1.applymap(labels.get)
sns.set_theme(style="ticks")
ax = sns.displot(x=q1, data=q1, hue='Alder', palette=sns.color_palette("husl", 3))
ax.set(xlabel='enkelt', ylabel='lelele')

plt.show()
```
